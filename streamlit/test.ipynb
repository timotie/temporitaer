```python
import pandas as pd
import glob
import folium
from folium.plugins import MarkerCluster
```

```python
path = r"source\\wea"

all_files = glob.glob(path + "/*.csv")
print(all_files)
li = []
for filename in all_files:
    df = pd.read_csv(filename,  on_bad_lines='skip', sep = ";", decimal= ".")
    li.append(df)
```

```python
dfAll = pd.concat(li, axis=0, ignore_index=True)
display(dfAll)
```

```python
dfAll["Koordinate: Breitengrad (WGS84)"] = dfAll["Koordinate: Breitengrad (WGS84)"].str.replace(',','.', regex= True).astype(float)
print(dfAll["Koordinate: Breitengrad (WGS84)"])
```

```python
m = folium.Map((51.24927460972442, 9.955969761190305), zoom_start= 5)
marker_cluster = MarkerCluster().add_to(m)
```

```python
for index, each in dfAll.iterrows():
    print((each["Koordinate: Breitengrad (WGS84)"]))
    folium.Marker((each["Koordinate: Breitengrad (WGS84)"], each["Koordinate: Längengrad (WGS84)"])).add_to(marker_cluster)
```
